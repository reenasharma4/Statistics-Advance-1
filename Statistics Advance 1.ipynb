# Statistics Advance 1

#1. What is a random variable in probability theory?
  - A random variable is a way to assign the numbers to get an outcomes of a random experiment. Instead of dealing with abstract events like rolling a dice, we use a function that gives a numerical result, making it easier to analyze the probability.

#2. What are the types of random variables?
  - There are two types of random variables:
    1. Discrete Random Variable: It takes specific and countable values. It can not take in between values. For example, number of heads in a coin toss.

    2. Continuous Random Variable: It can take any value within a given range. including decimals and fractions. It is not countable because there are infinitely many possible values. For example, height, weight, or time.

#3. What is the difference between discrete and continuous distributions?
  - The difference between the Discrete and Continuous Distributions are:
    1. Discrete distribution deals with individual values like rolling a die, where outcomes are 1, 2, 3, 4, 5, or 6.

    2. Continuous Distribution deals with an infinite number of values in a range like measuring rainfall in a city.

#4. What are probability distribution functions (PDF)?
  - Probability Distribution Function (PDF): A PDF describes that how probabilities are spread across possible values of a random variable.

  For discrete variables, it is called the Probability Mass Function (PMF) and assigns probabilities to each individual outcome. For continuous variables, the PDF gives the likelihood of values within an interval but not at a single point.

#5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - PDF tells how likely a value is in a small range and Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a certain value. It is like adding up all previous probabilities.

#6. What is a discrete uniform distribution?
  - A discrete uniform distribution means that all possible outcomes have the same probability. It applies when there are a fixed number of values and each is equally likely. Fox example: Rolling a fair 6-sided die → Outcomes (1, 2, 3, 4, 5, 6), each with probability 1/6.

#7. What are the key properties of a Bernoulli distribution?
  - A Bernoulli distribution represents a single experiment with only two possible outcomes: Success (1) (e.g., heads in a coin flip) and Failure (0) (e.g., tails in a coin flip).

  - Properties:
    1. The probability of success is p.
    2. The probability of failure is 1 - p.
    3. It is the simplest form of probability distribution.

#8. What is the binomial distribution, and how is it used in probability?
  - A binomial distribution models is the number of successes in multiple independent Bernoulli trials.
  - Formula: P(X=k)=(n/k)pk(1-p)n-k
     Where:
     n = number of trials
     k = number of successes
     p = probability of success
  - Used in: Quality control, A/B testing, and election predictions.

#9. What is the Poisson distribution and where is it applied?
  - A Poisson distribution models is the number of events happening in a fixed time or space when the events occur randomly and independently. For example: Number of calls a call center gets per hour, Number of customers arriving at a shop in 10 minutes.
  - Used in: Traffic analysis, website visits, and biology (mutation rates).

#10. What is a continuous uniform distribution?
  - A continuous uniform distribution means any value in a range is equally likely. For example: If a bus arrives anytime between 9:00 and 9:30, every minute in this period is equally probable.

#11. What are the characteristics of a normal distribution?
   - A normal distribution (bell curve) is a symmetric probability distribution where most values are clustered around the mean.
   - Key Characteristics are:
     1. Symmetrical around the mean (center)
     2. Mean = Median = Mode
     3. 68-95-99.7 Rule:
            68% of values fall within 1 standard deviation (σ) of the mean
            95% within 2σ
            99.7% within 3σ
     4. Skewness=0

#12. What is the standard normal distribution, and why is it important?
   - It is a continuous probability distribution. A special case of normal distribution is--> Mean=0, Standard deviation=1
   - Imporatnce of SND:
     1. It allows comparison of different datasets.
     2. It helps in Z-score calculations, which show how far a value is from the mean.
   
#13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
   - The Central Limit Theorem (CLT) states that if we taking a large enough random samples from any population, their averages will follow a normal distribution, regardless of the original distribution.
   - It is important because:
     1. It makes statistics work and allows us to use normal distribution for real-world data, even when the original data isn't normal.
     2. It enables confidence intervals and hypothesis testing.

#14. How does the Central Limit Theorem relate to the normal distribution?
   - Even if the original data is skewed or irregular, the sample will be normally distributed if the sample size is large enough. The larger the sample size, the closer the sampling distribution gets to a normal shape.
   - Fox example: If we measure the average height of students in many random samples, the distribution of these sample averages will be normal, even if individual heights are not.

#15. What is the application of Z statistics in hypothesis testing?
   - The Z-statistics (or Z-score) is used when there are the population variance is known and sample size is large (n ≥ 30).
   - It is uses in Hypothesis Testing:
     1. Checks whether a sample mean is significantly different from a population mean.
     2. Helps in determining a p-values, which tell if a result is statistically significant.
     3.Used in Z-tests for comparing means and proportions.

#16. How do you calculate a Z-score, and what does it represent?
   - A Z-score tells us about how far a particular data point is from the average (mean) in terms of standard deviations. It also helps us to understand whether a value is typical or unusual within a dataset.
   - The formula is: Z=X-μ/σ
     1. X = the value we are analyzing
     2. μ = the mean (average) of the dataset
     3. σ = the standard deviation
  - A Z-score of 0 means the value is exactly at the average.
  - A positive Z-score means the value is above the average.
  - A negative Z-score means the value is below the ave

#17. What are point estimates and interval estimates in statistics?
   - Point Estimate: A single number that estimates a population parameter.
     For Example: The average height of 50 randomly chosen people is 5.6 feet. That is a point estimate of the true population height.

   - Interval Estimate: A range of values that likely contains the true population parameter.
   For Example: Instead of just saying 5.6 feet, we say, We are 95% confident that the true average height is between 5.4 and 5.8 feet.

#18. What is the significance of confidence intervals in statistical analysis?
   - A confidence interval gives us a range where we expect the true population value to be. It helps account for uncertainty in data collection.
   - It is important because it helps researchers avoid making conclusions based on incomplete data, a narrower confidence interval means a more precise estimate and a wider confidence interval means more uncertainty in the estimate.

#19. What is the relationship between a Z-score and a confidence interval?
   - A Z-score shows how far a value is from the mean in terms of standard deviations. Confidence intervals use Z-scores to set a range where the true population mean is likely to be. For example, a 95% confidence interval uses a Z-score of 1.96 to determine the margin of error.
   
#20. How are Z-scores used to compare different distribution?
   - Z-scores standardize values across different data sets, allowing a fair comparisons. Even if two distributions have different means and standard deviations, Z-scores help us to compare individual performances relative to their respective groups.

#21. What are the assumptions for applying the Central Limit Theorem?
   - The CLT states that large random samples (n ≥ 30) from any population will have a normal distribution of sample means.
   - Assumptions are:
     1. Large sample size (n ≥ 30) → Works better with bigger samples.
     2. Random & Independent samples → No bias, one sample does not affect another.
     3. Finite variance → No extreme values.

#22. What is the concept of expected value in a probability distribution?
   - The expected value is the long-term average outcome of a random event.
   - Formula: E(X):∑[X*P(X)]
   - Example: Rolling a fair die → Average outcome = 3.5 over time.

#23. How does a probability distribution relate to the expected outcome of a random variable?
   - A probability distribution shows all possible outcomes & their chances. The expected value is the average result over many tries.
   - Example: Coin Toss (Heads = $5, Tails = $0, 50% chance each), (Expected earnings per toss = $2.50)

```python
#4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

n, p = 10, 0.5
data = binom.rvs(n, p, size=1000)

plt.hist(data, bins=10, density=True, alpha=0.6, color='b', edgecolor='black')
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import scipy.stats as stats

# Function to calculate Z-score
def z_score(value, mean, std):
    return (value - mean) / std

# Example
data_point = 85
mean = 70
std = 10

z = z_score(data_point, mean, std)
p_value = stats.norm.cdf(z)

print(f"Z-score: {z}")
print(f"Probability of getting value ≤ {data_point}: {p_value}")
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.
import scipy.stats as stats
import numpy as np

# Sample data
sample = [65, 70, 75, 80, 85]
sample_mean = np.mean(sample)
pop_mean = 72  # Population mean
std_dev = 10
n = len(sample)

# Compute Z-score
z = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = stats.norm.sf(abs(z)) * 2  # Two-tailed test

print(f"Z-score: {z}, P-value: {p_value}")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result.
import scipy.stats as stats
import numpy as np

# Sample data
data = [12, 15, 14, 10, 18, 20, 13]
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# Confidence Interval (95%)
ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std/np.sqrt(n))

print(f"95% Confidence Interval: {ci}")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

# Generate normal data
data = np.random.normal(50, 10, 100)
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# Confidence Interval (95%)
ci = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(n))
print(f"Confidence Interval: {ci}")
```
